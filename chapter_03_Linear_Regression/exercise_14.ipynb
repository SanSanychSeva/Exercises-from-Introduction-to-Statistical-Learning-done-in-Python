```python
import datetime as dt
start_run = dt.datetime.now()

import math
import pandas as pd
import numpy as np
import scipy.stats as statmath 
from sklearn import linear_model
#from sklearn.metrics import mean_squared_error, r2_score

import seaborn as sns
import matplotlib.pyplot as plt
```

# Functions for used statistical formulas

```python
def p_value(t_stat, len_df):
    return statmath.t.sf(abs(t_stat), len_df)*2
```

# items a: data generation

```python
n_sampling = 100                                             # sampling size paramer for what-if-analysis
rng = np.random.default_rng()                                # creating Numpy-generator object
X1 = rng.uniform(size=n_sampling)                            # X1-vector with uniform-distribution over [0;1]
X2 = X1 / 2 + rng.standard_normal(size=n_sampling) / 10      # X2-vector with stochastized collinearity to X1
Y = 2 + 2*X1 + 0.3*X2 + rng.standard_normal(size=n_sampling) # Y-vector, with true nature of f(x): y = 2+2*x1+0.3*x2


df_YontoX = pd.DataFrame(
    [X1,X2,Y],
    index=['X1','X2','Y']
).T

df_YontoX.head()
```

## Conclusions:
1. Generated data correspond to the linear model

$$ f(x)\, :\,\, y = 2 + 2x_1 + 0.3x_2 $$

2. The regression coefficients of this model are the true values behind them as:

$$\begin{eqnarray}
\beta_0 &=& 2 &; \\
\beta_1 &=& 2 &; \\
\beta_2 &=& 0.3 &. \\
\end{eqnarray}$$

# item c: multiple linear regression of `Y` onto `X` (MLR):

```python
X = np.array(df_YontoX.drop(columns='Y'))

MLR_model_obj_YontoX = linear_model.LinearRegression()   # creating Sklearn-object for MLR-model
MLR_model_obj_YontoX.fit(X,Y)                            # training MLR-model on the generated dataset

# let's start building a summary table for the results for easy comparisons:

params_tbl = pd.DataFrame(index=['R2_score','intercept','coef_x1','coef_x2'])
params_tbl['true_f(x)'] = [1.0,2.0,2.0,0.3]

# adding model fitting results to it:

params_tbl['smpl_MLR'] = [
    MLR_model_obj_YontoX.score(X,Y),
    MLR_model_obj_YontoX.intercept_,
    MLR_model_obj_YontoX.coef_[0],
    MLR_model_obj_YontoX.coef_[1],
]

params_tbl                            #show table
```

well, already the $ \hat\beta_1 \,\mbox{and}\, \hat\beta_2 $ looks very disgusting as compared to $ \beta_1 \,\mbox{and}\, \beta_2 $ !  

Let's estimate their `std` and $ H_0 \, :\,\, \beta_i = 0 $ `t-statistics` and `p-values`.  As soon as our library function does not generate it as model parameters (unlike in R) we need either to make bootstrapping, or in our case (as the data are generated by us) we can simply repeat the data generation and model fitting for each new sample to collect their statistics directly:

```python
bootstrap_df = pd.DataFrame(columns=['R2_score','intercept','coef_x1','coef_x2'])

MLR_model_obj_YontoX_bs = linear_model.LinearRegression()   # creating different Sklearn-object for MLR-model

for i in range(n_sampling):
    
    X1_bs = rng.uniform(size=n_sampling)                            
    X2_bs = X1_bs / 2 + rng.standard_normal(size=n_sampling) / 10   
    Y_bs = 2 + 2*X1_bs + 0.3*X2_bs + rng.standard_normal(size=n_sampling) 

    df_YontoX_bs = pd.DataFrame(
        [X1_bs,X2_bs,Y_bs],
        index=['X1','X2','Y']
    ).T
    X_bs = np.array(df_YontoX_bs.drop(columns='Y'))
    
    MLR_model_obj_YontoX_bs.fit(X_bs,Y_bs)    
    
    bootstrap_df.loc[i] = [
        MLR_model_obj_YontoX_bs.score(X_bs,Y_bs),
        MLR_model_obj_YontoX_bs.intercept_,
        MLR_model_obj_YontoX_bs.coef_[0],
        MLR_model_obj_YontoX_bs.coef_[1]
    ]

bootstrap_df.describe().drop(index=['25%','50%','75%'])
```

let's add into parameters comparisons table

```python
params_tbl = params_tbl.join(bootstrap_df.describe().T[['std','mean']].rename(columns={
    'std':'std_MLR',
    'mean':'mean_MLR'
}))

params_tbl['L_conf_MLR'] = params_tbl['mean_MLR'] - params_tbl['std_MLR']*2/math.sqrt(n_sampling)
params_tbl['R_conf_MLR'] = params_tbl['mean_MLR'] + params_tbl['std_MLR']*2/math.sqrt(n_sampling)

params_tbl
```

## Hypothesis checking
### $H_0\, :\,\, \beta_i = 0$

```python
df_tmp = params_tbl.drop(index='R2_score')
df_tmp['t-stat'] = df_tmp['smpl_MLR'] / df_tmp['std_MLR']
df_tmp['p-value'] = df_tmp['t-stat'].apply(lambda t_stat: p_value(t_stat, n_sampling))

show_MLR = df_tmp
round(show_MLR,5)
```

### $H_0\, :\,\, \beta_i = \beta_i^{true}$

```python
df_tmp = params_tbl.drop(index='R2_score')
df_tmp['t-stat'] = (df_tmp['smpl_MLR'] - df_tmp['true_f(x)'] ) / df_tmp['std_MLR']
df_tmp['p-value'] = df_tmp['t-stat'].apply(lambda t_stat: p_value(t_stat, n_sampling))

def check_f(df):
    if df['L_conf_MLR'] <= df['true_f(x)'] <= df['R_conf_MLR']:
        return True
    else:
        return False

df_tmp['if true Beta fits conf.int. for mean sampling?'] = df_tmp.apply(check_f,axis=1)

show_MLR_bis = df_tmp
round(show_MLR_bis,5)
```

## Conclusions:
1. The $H_0\, :\,\, \beta_i = 0$ can be rejected only for $ \beta_0\, \mbox{and}\, \beta_1 $, but not for $\beta_2$ .
2. The $H_0\, :\,\, \beta_i = \beta_i^{true}$ cannot be rejected for all $\beta_i$, yet the confidence interval for mean sampling does fit the $\beta_i^{true}$ .

# item d: simple linear regression of `Y` onto `X1` (SLR_x1)

```python
SLR_model_obj_YontoX1 = linear_model.LinearRegression()   
SLR_model_obj_YontoX1.fit(X1.reshape(-1, 1),Y)            

params_tbl['smpl_SLR_x1'] = [
    SLR_model_obj_YontoX1.score(X1.reshape(-1, 1),Y),
    SLR_model_obj_YontoX1.intercept_,
    SLR_model_obj_YontoX1.coef_[0],
    None,
]          

bootstrap_df = pd.DataFrame(columns=['R2_score','intercept','coef_x1','coef_x2'])

SLR_model_obj_YontoX1_bs = linear_model.LinearRegression()   

for i in range(n_sampling):
    
    X1_bs = rng.uniform(size=n_sampling)                            
    X2_bs = X1_bs / 2 + rng.standard_normal(size=n_sampling) / 10   
    Y_bs = 2 + 2*X1_bs + 0.3*X2_bs + rng.standard_normal(size=n_sampling) 

    SLR_model_obj_YontoX1_bs.fit(X1_bs.reshape(-1, 1),Y_bs)    
    
    bootstrap_df.loc[i] = [
        SLR_model_obj_YontoX1_bs.score(X1_bs.reshape(-1, 1),Y_bs),
        SLR_model_obj_YontoX1_bs.intercept_,
        SLR_model_obj_YontoX1_bs.coef_[0],
        None
    ]

params_tbl = params_tbl.join(bootstrap_df.describe().T[['std','mean']].rename(columns={
    'std':'std_SLR_x1',
    'mean':'mean_SLR_x1'
}))

params_tbl[['true_f(x)','smpl_MLR','smpl_SLR_x1','std_MLR','std_SLR_x1','mean_MLR','mean_SLR_x1']]
```

## Hypothesis checking

### $H_0\, :\,\, \beta_i = 0$

```python
df_tmp = params_tbl.drop(index=['R2_score','coef_x2'])
df_tmp['t-stat'] = df_tmp['smpl_SLR_x1'] / df_tmp['std_SLR_x1']
df_tmp['p-value'] = df_tmp['t-stat'].apply(lambda t_stat: p_value(t_stat, n_sampling))

show_SLR_x1 = df_tmp

print('\nfor Y regression onto X1 and X2:', end=' ')
display(show_MLR[['t-stat','p-value']].iloc[[0,1]])
print('\nfor Y regression onto X1 only:', end=' ')
display(show_SLR_x1[['t-stat','p-value']])
```

### $H_0\, :\,\, \beta_i = \beta_i^{true}$

```python
df_tmp = params_tbl.drop(index=['R2_score','coef_x2'])
df_tmp['t-stat'] = ( df_tmp['smpl_SLR_x1'] - df_tmp['true_f(x)'] ) / df_tmp['std_SLR_x1']
df_tmp['p-value'] = df_tmp['t-stat'].apply(lambda t_stat: p_value(t_stat, n_sampling))

show_SLR_x1_bis = df_tmp

print('\nfor Y regression onto X1 and X2:', end=' ')
display(show_MLR_bis[['t-stat','p-value']].iloc[[0,1]])
print('\nfor Y regression onto X1 only:', end=' ')
display(show_SLR_x1_bis[['t-stat','p-value']])
```

## Conclusions:
1. For $\beta_1$ `std_SLR_x1` was almost twice less than `std_MLR`
2. Naturally for the hypothesis $ H_0\, :\,\, \beta_1 = 0 $ the `t-statistics` gets twice
3. And yet somehow for the hypothesis $ H_0\, :\,\, \beta_1 = 2 $ the `t-statistics` is not very impressive - why?
4. This can be explained like that:
 * when we drop X2, because of collenearity we must use as the true $ \beta_1^{true}\, =\, 2 + 0.3*0.5 - \omicron_{st} \, =\, 2.15 - \omicron_{st} $ , where $\omicron_{st}$ is actually the *small* influence of stochastisity introduced in X2, as investigated in earlier exercises to this chapter.
 
Let's check the last one:

```python
df_tmp = params_tbl.drop(index=['R2_score','coef_x2','intercept'])
df_tmp['t-stat'] = ( df_tmp['smpl_SLR_x1'] - 2.15 ) / df_tmp['std_SLR_x1']
df_tmp['p-value'] = df_tmp['t-stat'].apply(lambda t_stat: p_value(t_stat, n_sampling))

print('\nfor Ho: beta_1=2,15', end=' ')
df_tmp[['t-stat','p-value']]
```

not enough good, but if we use mean value instead of sample value:

```python
df_tmp = params_tbl.drop(index=['R2_score','coef_x2','intercept'])
df_tmp['t-stat'] = ( df_tmp['mean_SLR_x1'] - 2.15 ) / df_tmp['std_SLR_x1']
df_tmp['p-value'] = df_tmp['t-stat'].apply(lambda t_stat: p_value(t_stat, n_sampling))

print('\nfor Ho: beta_1=2,15', end=' ')
df_tmp[['t-stat','p-value']]
```

now good - the mean $\beta_1$ is very close to 2.15 !

# item e: simple linear regression of `Y` onto `X2` (SLR_x2)

```python
SLR_model_obj_YontoX2 = linear_model.LinearRegression()   
SLR_model_obj_YontoX2.fit(X2.reshape(-1, 1),Y)            

params_tbl['smpl_SLR_x2'] = [
    SLR_model_obj_YontoX2.score(X2.reshape(-1, 1),Y),
    SLR_model_obj_YontoX2.intercept_,
    None,
    SLR_model_obj_YontoX2.coef_[0],
]          

bootstrap_df = pd.DataFrame(columns=['R2_score','intercept','coef_x1','coef_x2'])

SLR_model_obj_YontoX2_bs = linear_model.LinearRegression()   

for i in range(n_sampling):
    
    X1_bs = rng.uniform(size=n_sampling)                            
    X2_bs = X1_bs / 2 + rng.standard_normal(size=n_sampling) / 10   
    Y_bs = 2 + 2*X1_bs + 0.3*X2_bs + rng.standard_normal(size=n_sampling) 

    SLR_model_obj_YontoX2_bs.fit(X2_bs.reshape(-1, 1),Y_bs)    
    
    bootstrap_df.loc[i] = [
        SLR_model_obj_YontoX2_bs.score(X2_bs.reshape(-1, 1),Y_bs),
        SLR_model_obj_YontoX2_bs.intercept_,
        None,
        SLR_model_obj_YontoX2_bs.coef_[0],
    ]

params_tbl = params_tbl.join(bootstrap_df.describe().T[['std','mean']].rename(columns={
    'std':'std_SLR_x2',
    'mean':'mean_SLR_x2'
}))

params_tbl[['true_f(x)','smpl_MLR','smpl_SLR_x1','smpl_SLR_x2','std_MLR','std_SLR_x1','std_SLR_x2',
            'mean_MLR','mean_SLR_x1','mean_SLR_x2']]
```

## Hypothesis checking

### $H_0\, :\,\, \beta_i = 0$

```python
df_tmp = params_tbl.drop(index=['R2_score','coef_x1'])
df_tmp['t-stat'] = df_tmp['smpl_SLR_x2'] / df_tmp['std_SLR_x2']
df_tmp['p-value'] = df_tmp['t-stat'].apply(lambda t_stat: p_value(t_stat, n_sampling))

show_SLR_x2 = df_tmp

print('\nfor Y regression onto X1 and X2:', end=' ')
display(show_MLR[['t-stat','p-value']].iloc[[0,2]])
print('\nfor Y regression onto X2 only:', end=' ')
display(show_SLR_x2[['t-stat','p-value']])
```

### $H_0\, :\,\, \beta_i = \beta_i^{true}$

## Conclusions:
1. For $\beta_2$ `std_SLR_x2` was more than twice less than `std_MLR`.
2. Naturally for the hypothesis $ H_0\, :\,\, \beta_2 = 0 $ the `t-statistics` gains a lot.
3. As well as for the hypothesis $ H_0\, :\,\, \beta_2 = 0.3 $.
4. So in both case the $H_0$ can be rejected
5. This can be explained like that:
 * when we drop X1, because of collenearity we must use as the true $ \beta_1^{true}\, =\, 0.3+ 2*2 - O_{st} \, =\, 4.3 - O_{st} $ , where $O_{st}$ is actually the *huge* influence of stochastisity introduced in X2, as investigated in earlier exercises to this chapter.
6. As we do not know $O_{st}$ we cannot check the last one this time exactly, but we can use mean $\beta_2 = 3.056$ as a good approximation of $\beta_1^{true}$.
7. Thus we can estimate the stochastic influence on $\beta_2$ in Y regression on X2 only as 4.3 - 3.1 = 1.2 - the reasons were investigated in earlier exercises of this chapter - it is because X2 has stochastic term so the slope is heavily pulled down from 4.3 to 3.1.

2. This instability is valid not only to one sample coefficients, but also to their mean values!
3. Therefore we can conclude, that in case of collenearity of predictors the main benefit of linear model (interpretability of coefficients) can be easily lost by only one occasional outliner added to the data set by some mistake!
4. But alas, in reality we cannot say which samples are following the hidden f(x) and which are irrelevant to it, **so the only way to keep interpretability of linear model is to be sure that there are no collinearity among the predictors**.
5. Alas the binary correlator form can help only in case dim X = 2, if dim X > 2, there could be hidden collinearity when X1 = a * X2 + b * X3.  As an idea for dim X > 2 we can put aside Y and try to predict by MLR all Xi our of X - in case the X1 = a * X2 + b * X3 is behind the data, we should get R2 for prediction of X1, X2 and X3 anomally high.
6. Despite unstable coefficients for MLR in case of collinearity of predictors, it does not obligatory mean that prediction of Y shall be bad - simply we can get close predictions for very different set of betas!  So only interpretability shall really suffer.
