# item c: Multiple Linear Regression of `mpg`

```python
X = np.array(df.drop(columns=['mpg','name']) )
Y = np.array(df['mpg'])

MLR_model_obj = linear_model.LinearRegression() # creating model object
MLR_model_obj.fit(X,Y)

def mpg_f(X_test): return MLR_model_obj.predict(X_test)
```

## precision of model on training data

```python
print('\nMean squared error:', MSE_0 := mean_squared_error(Y, MLR_model_obj.predict(X)), '\n' )

print('Coefficient of determination - R^2 statistics (variant 1):', R2_0 := MLR_model_obj.score(X,Y) )
print('Coefficient of determination - R^2 statistics (variant 2):', r2_score(Y, MLR_model_obj.predict(X)) )
```

## t_statistics for each parameter
alas, there is no params like t-statistics (or f-) in the library model - so we might go for some bootstrapping:

```python
def random_one_half_df(df, rng = np.random.default_rng()):  # random sampling for 1/2 rows from df
    n = len(df)//2
    rows_list = []
    while len(rows_list) < n:
        candidate = rng.integers(len(df))
        if candidate not in rows_list:  rows_list.append(candidate)
    return df.loc[rows_list]

sample_MLR_model_obj = linear_model.LinearRegression() # creating model object
beta = []                                              # here we save MLR parameters for random samples

bootstrap_n = 2*len(df)                                # the more - the better, yet it is getting long
for i in range(bootstrap_n):                     
    sample_df = random_one_half_df(df)
    x = np.array(sample_df.drop(columns=['mpg','name']))
    y = np.array(sample_df['mpg'])
    sample_MLR_model_obj.fit(x,y)
    beta.append([sample_MLR_model_obj.intercept_] + list(sample_MLR_model_obj.coef_))
    
MLR_stat_df_samples = pd.DataFrame(
    beta,
    columns = ['intercept']+list(df.drop(columns=['mpg','name']).columns)
)
```

```python
print("\nLet's compare coefficients between model and bootstrapping-mean:")

confidence_df = MLR_stat_df_samples.describe().iloc[[0,1,2]].T

confidence_df['model'] = [ MLR_model_obj.intercept_] + list(MLR_model_obj.coef_)

confidence_df['delta_mean_to_model'] = confidence_df['mean'] - confidence_df['model']

confidence_df['delta_vs_std/sqroot(N)'] = (confidence_df['delta_mean_to_model'] / confidence_df['std'] * 
                                           math.sqrt(len(df)) ).map(abs).map(lambda x: round(x,1))

confidence_df['t_stat_v1'] = confidence_df['mean'] / (confidence_df['std'] / confidence_df['count'].map(math.sqrt) )

confidence_df['p_value_v1'] = confidence_df['t_stat_v1'].map(lambda x: statmath.t.sf(abs(x), len(df)//2 - 1) *2)

confidence_df['(t_stat, p_value)_v2'] = [ statmath.ttest_1samp(list(MLR_stat_df_samples[bt]),0) 
                                          for bt in list(confidence_df.index) ]

round(confidence_df,5)
```

# item d: Outliers
we shall use the leverage-residual studentised coords to identify ourliers:
* `horisontal axis (leverage):` $Y_{predicted}\ -\ \bar{Y}_{predicted}$ / studentized  (actually for linear prediction $\bar{Y}_{predicted}\ =\ \bar{Y}_{actual}$)
* `vertical axis   (residual):` $Y_{actual}\ -\ Y_{predicted}$ / studentized

```python
# adding predicted values to main dataframe
df['mpg_pred'] = MLR_model_obj.predict(X)
```

```python
print('check that mean value of average predicted value is equal to the mean value: relative_delta=', 
      (df['mpg_pred'].mean() - df['mpg'].mean())/df['mpg'].mean())
```

```python
# adding horisontal coord for outliers analyses to main dataframe
df['fitted_value_studentised'] = df['mpg_pred'] - df['mpg_pred'].mean()
df['fitted_value_studentised'] = df['fitted_value_studentised'] / df['fitted_value_studentised'].std()
```

```python
# adding vertical coord for outliers analyses to main dataframe
df['residual_value_studentized'] = df['mpg'] - df['mpg_pred']
df['residual_value_studentized'] = df['residual_value_studentized'] / df['residual_value_studentized'].std()
```

```python
show_df = df.sort_values('fitted_value_studentised')
show_df['smooth_fit_residuals'] = show_df['residual_value_studentized'].rolling(3).mean(
                                                                      ).rolling(5).mean(
                                                                      ).rolling(7).mean(
                                                                      ).rolling(11).mean(
                                                                      ).rolling(13).mean()
```

```python
plt.figure(figsize=(12,7))

sns.scatterplot(data=show_df, x='fitted_value_studentised', y='residual_value_studentized', marker='x')
sns.lineplot(data=show_df, x='fitted_value_studentised', y='smooth_fit_residuals', color='red', 
             label='smooth fit to residuals')
plt.axhline(y=0, color='black', label='0 - predicted mpg value level')

plt.xlim(-3,3)
plt.ylim(-4,4)
plt.legend()
plt.grid()
plt.show()
```

## Conclusions:
1. Different swaps for positive and negative residuals is caused by nonelinear shape of the true f(X):
 * in the very middle of distibution prediction is above actual mpg (|leverage| < 1 std),
 * on the boundary of distribution prediction is less then actual (|leverage| > 1 std)
3. There are no big outliers worth of deleting, except maybe right top corner of the distribution.

## yet let's play with dropping outliers:

```python
X_good = np.array(df.query('abs(residual_value_studentized) < 2.5 & abs(fitted_value_studentised) < 2').drop(
    columns=['mpg','name','residual_value_studentized','fitted_value_studentised']) )
Y_good = np.array(df.query('abs(residual_value_studentized) < 2.5 & abs(fitted_value_studentised) < 2')['mpg'])

good_MLR_model_obj = linear_model.LinearRegression() # creating model object
good_MLR_model_obj.fit(X_good,Y_good)

print('\nHAVING DROPPED',
      N_3 := len(df)-len(df.query('abs(residual_value_studentized) < 2.5 & abs(fitted_value_studentised) < 2')),
      'OUTLIERS OUT OF', len(df),'WE HAVE SOME IMPROVEMENT IN PREDICTION ON TRAINING DATA:')
print('\nMean squared error:', MSE_3 := mean_squared_error(Y_good, good_MLR_model_obj.predict(X_good)), '\n' )

print('Coefficient of determination - R^2 statistics (variant 1):', R2_3 := good_MLR_model_obj.score(X_good,Y_good) )
print('Coefficient of determination - R^2 statistics (variant 2):', r2_score(Y_good, good_MLR_model_obj.predict(X_good)) )
```

### if we drop only high leverage outliers:

```python
X_good = np.array(df.query('abs(fitted_value_studentised) < 2').drop(
    columns=['mpg','name','residual_value_studentized','fitted_value_studentised']) )
Y_good = np.array(df.query('abs(fitted_value_studentised) < 2')['mpg'])

good_MLR_model_obj = linear_model.LinearRegression() # creating model object
good_MLR_model_obj.fit(X_good,Y_good)

print('\nHAVING DROPPED', N_1 := len(df)-len(df.query('abs(fitted_value_studentised) < 2')),'OUTLIERS OUT OF',
      len(df),'WE SHOULD INFLUENCE MAINLY R^2 STATISTICS:')
print('\nMean squared error:', MSE_1 := mean_squared_error(Y_good, good_MLR_model_obj.predict(X_good)), '\n' )

print('Coefficient of determination - R^2 statistics (variant 1):', R2_1 := good_MLR_model_obj.score(X_good,Y_good) )
print('Coefficient of determination - R^2 statistics (variant 2):', r2_score(Y_good, good_MLR_model_obj.predict(X_good)) )
```

### if we drop only high residual outliers:

```python
X_good = np.array(df.query('abs(residual_value_studentized) < 2.5').drop(
    columns=['mpg','name','residual_value_studentized','fitted_value_studentised']) )
Y_good = np.array(df.query('abs(residual_value_studentized) < 2.5')['mpg'])

good_MLR_model_obj = linear_model.LinearRegression() # creating model object
good_MLR_model_obj.fit(X_good,Y_good)

print('\nHAVING DROPPED', N_2 := len(df)-len(df.query('abs(residual_value_studentized) < 2.5')),'OUTLIERS OUT OF',
      len(df),'WE SHOULD MAINLY IMPROVE MSE:')
print('\nMean squared error:', MSE_2 := mean_squared_error(Y_good, good_MLR_model_obj.predict(X_good)), '\n' )

print('Coefficient of determination - R^2 statistics (variant 1):', R2_2 := good_MLR_model_obj.score(X_good,Y_good) )
print('Coefficient of determination - R^2 statistics (variant 2):', r2_score(Y_good, good_MLR_model_obj.predict(X_good)) )
```

### Conclusion: so, high leverage outliers are no problem if their residual is small!

### hense let's try optimised outliers-drop criteria:

```python
cut = 8 # tuning threshold level

X_good = np.array(df.query(
    'residual_value_studentized*residual_value_studentized+fitted_value_studentised*fitted_value_studentised < @cut'
).drop(
    columns=['mpg','name','residual_value_studentized','fitted_value_studentised']) )
Y_good = np.array(df.query(
    'residual_value_studentized*residual_value_studentized+fitted_value_studentised*fitted_value_studentised < @cut'
)['mpg'])

good_MLR_model_obj = linear_model.LinearRegression() # creating model object
good_MLR_model_obj.fit(X_good,Y_good)

print('\nHAVING DROPPED', N_4 := len(df)-len(df.query(
      'residual_value_studentized*residual_value_studentized+fitted_value_studentised*fitted_value_studentised < @cut'
      )),'OUTLIERS OUT OF',
      len(df),'WE DO NOT HAVE YET BIGGEST IMPROVEMENT IN PREDICTION ON TRAINING DATA:')
print('\nMean squared error:', MSE_4 := mean_squared_error(Y_good, good_MLR_model_obj.predict(X_good)), '\n' )

print('Coefficient of determination - R^2 statistics (variant 1):', R2_4 := good_MLR_model_obj.score(X_good,Y_good) )
print('Coefficient of determination - R^2 statistics (variant 2):', r2_score(Y_good, good_MLR_model_obj.predict(X_good)) )
```

# item e: Small nonelinearity of f(X)
We need to add quadratic terms, but there are plenty of them!

```python
df = pd.read_csv('datasets/Auto.csv')
df = df[df['horsepower'].map(str.isdecimal)]
df['horsepower'] = pd.to_numeric(df['horsepower'])
df = df.reset_index(drop=True).drop(columns='name')
df.info()
df.head(2)
```

```python
clist = list(df.columns)
n = len(clist)

for i in range(1,n):
    for j in range(i,n):
        c1 = clist[i]
        c2 = clist[j]
        df[c1+'_'+c2] = df[c1]*df[c2]

df.info()
```

```python
quadroX = np.array(df.drop(columns='mpg') )
Y = np.array(df['mpg'])

quadroMR_model_obj = linear_model.LinearRegression() # creating model object
quadroMR_model_obj.fit(quadroX,Y)

print('\nMean squared error:', mean_squared_error(Y, quadroMR_model_obj.predict(quadroX)), '\n' )

print('Coefficient of determination - R^2 statistics (variant 1):', quadroMR_model_obj.score(quadroX,Y) )
print('Coefficient of determination - R^2 statistics (variant 2):', r2_score(Y, quadroMR_model_obj.predict(quadroX)) )
```

```python
def random_one_half_df(df, rng = np.random.default_rng()):  # random sampling for 1/2 rows from df
    n = len(df)//2
    rows_list = []
    while len(rows_list) < n:
        candidate = rng.integers(len(df))
        if candidate not in rows_list:  rows_list.append(candidate)
    return df.loc[rows_list]

sample_quadroMR_model_obj = linear_model.LinearRegression() # creating model object
beta = []                                              # here we save quadroMR parameters for random samples

bootstrap_n = 2*len(df)                                # the more - the better, yet it is getting long
for i in range(bootstrap_n):                     
    sample_df = random_one_half_df(df)
    x = np.array(sample_df.drop(columns='mpg'))
    y = np.array(sample_df['mpg'])
    sample_quadroMR_model_obj.fit(x,y)
    beta.append([sample_quadroMR_model_obj.intercept_] + list(sample_quadroMR_model_obj.coef_))
    
quadroMR_stat_df_samples = pd.DataFrame(
    beta,
    columns = ['intercept']+list(df.drop(columns='mpg').columns)
)
```

```python
print("\nLet's compare coefficients between model and bootstrapping-mean:")

confidence_df = quadroMR_stat_df_samples.describe().iloc[[0,1,2]].T

confidence_df['model'] = [ quadroMR_model_obj.intercept_] + list(quadroMR_model_obj.coef_)

confidence_df['delta_mean_to_model'] = confidence_df['mean'] - confidence_df['model']

confidence_df['delta_vs_std/sqroot(N)'] = (confidence_df['delta_mean_to_model'] / confidence_df['std'] * 
                                           math.sqrt(len(df)) ).map(abs).map(lambda x: round(x,1))

confidence_df['t_stat_v1'] = confidence_df['mean'] / (confidence_df['std'] / confidence_df['count'].map(math.sqrt) )

confidence_df['p_value_v1'] = confidence_df['t_stat_v1'].map(lambda x: statmath.t.sf(abs(x), len(df)//2 - 1) *2)

confidence_df['(t_stat, p_value)_v2'] = [ statmath.ttest_1samp(list(quadroMR_stat_df_samples[bt]),0) 
                                          for bt in list(confidence_df.index) ]

round(confidence_df.sort_values('t_stat_v1').drop(columns=['delta_mean_to_model','delta_vs_std/sqroot(N)']),5)
```

## Conclusions:
1. Adding quadratic interactions has greatly improved the precision of the model - ON TRAINING DATA!
2. **Although we have now too many degree of freedom (35) for only 392 samples, so only 11 samples per one degree of freedom - too few, we would need at least 30 for reliable statistics!**  And this can be seen by next item - if we repeat sampling, we get in item 3 as a results the two, three, four and even five lines of insignificant interactions, so p_value are not even stable as for sampling we used half dataset - with 5 independent samples per one degree of freedom!  This was **`the curse of dimensionality`**!
3. All quadratic interactions seems statistically significant except the ones with p_value > significance_alpha:

```python
alpha = 0.01
print('If significance level is taken as',alpha, 'than the only insignificant interactions are between:')
round(confidence_df.query('p_value_v1 > @alpha')[['t_stat_v1','p_value_v1']],5)
```

# item f: Using none linear trasformation of parameter values
Let's try to predict 1/mpg - as this is expected to be more linear towards horsepower!

```python
df = pd.read_csv('datasets/Auto.csv')
df = df[df['horsepower'].map(str.isdecimal)]
df['horsepower'] = pd.to_numeric(df['horsepower'])
df['mpg'] = 1 / df['mpg']
df = df.reset_index(drop=True)
df.info()
df.head(2)
```
