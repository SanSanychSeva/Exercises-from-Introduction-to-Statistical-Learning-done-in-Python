# Text of exercise 11 chapter 3 ISLRv2 book

In this problem we will investigate the `t-statistic` for the null hypothesis $H_0\ : \beta = 0$ in simple linear regression without an intercept. To begin, we generate a predictor `x` and a response `y` as follows.
* `set.seed (1)`
* `x <- rnorm (100)`
* `y <- 2 * x + rnorm (100)`

(a) Perform a simple linear regression of **`y` onto `x`**, *without* an intercept. Report the coefficient estimate $\cap{\beta}$, the `standard error` of this coefficient estimate, and the `t-statistic` and `p-value` associated with the null hypothesis $H_0\ \: \beta = 0$. Comment on these results. (You can perform regression without an intercept using
the command `lm(yâˆ¼x+0)`.)

(b) Now perform a simple linear regression of **`x` onto `y`** *without* an intercept, and report the coefficient estimate, its `standard error`, and the corresponding `t-statistic` and `p-values` associated with the null hypothesis $H_0\ \  : \beta = 0$. Comment on these results.

(c) What is the relationship between the results obtained in (a) and (b)?

(d) For the regression of Y onto X *without* an intercept, the `t-statistic` for $H_0\ : \beta = 0$ takes the form 

$$ \mbox{t-stat_v1}\ =\ \hat{\beta}\ /\ \mbox{SE}(\hat{\beta}), $$ 
where (given by (3.38)):
$$ \hat{\beta}\ =\ \frac
{\sum_{i=1}^{n} x_i y_i}
{\sum_{i'=1}^{n} x_{i'}^2}
,$$ 
and where:
$$ \mbox{SE}(\hat{\beta})\ =\ \sqrt{\frac
{\sum_{i=1}^{n} (y_i-x_i\hat{\beta})^2 }
{(n-1)\ \sum_{i'=1}^{n} x_{i'}^2}
}.$$

(These formulas are slightly different from those given in Sections 3.1.1 and 3.1.2, since here we are performing regression without an intercept.) Show algebraically, and confirm numerically in R, that the `t-statistic` can be written in symmetrical form as:

$$ \mbox{t-stat_v2}\ =\ \frac
{\sqrt{n-1}\ \sum_{i=1}^{n} x_i y_i}
{\sqrt{
\Bigl(\sum_{i'=1}^{n} x_{i'}^2\Bigr)\ \Bigl(\sum_{i''=1}^{n} y_{i''}^2\Bigr)\ -\ \Bigl(\sum_{i=1}^{n} x_i y_i\Bigr)^2
}}.$$

(e) Using the results from (d), argue that the `t-statistic` for the regression of **`y` onto `x`** is the same as the `t-statistic` for the regression of **`x` onto `y`**.

(f) In R, show that when regression is performed with an intercept, the `t-statistic` for $H_0\ : \beta_1 = 0$ is the same for the regression of **`y` onto `x`** as it is for the regression of **`x` onto `y`**.

\*) **ISLRv2** stands for: **An Introduction to Statistical Learning book by Gareth James, Daniela Witten, Trevor Hastie, Robert Tibshirani**

# Functions for used statistical formulas

```python
def beta_approx(df):
    return ( df['X'] * df['Y'] ).sum() / ( df['X'] * df['X'] ).sum()

def std_beta_approx(df):
    beta = beta_approx(df)
    disp = ( (df['Y'] - df['X']*beta) * (df['Y'] - df['X']*beta) ).sum(
             ) / ( df['X'] * df['X'] ).sum() / (len(df)-1)
    return math.sqrt(disp)

def t_stat_v1(df):
    return beta_approx(df) / std_beta_approx(df)

def t_stat_v2(df):
    top = ( df['X'] * df['Y'] ).sum() * math.sqrt(len(df)-1)
    bottom_sq = (
        (( df['X'] * df['X'] ).sum()) * (( df['Y'] * df['Y'] ).sum()) - 
        (( df['X'] * df['Y'] ).sum()) * (( df['X'] * df['Y'] ).sum())
    )
    return top / math.sqrt(bottom_sq)

def p_value(t_stat, len_df):
    return statmath.t.sf(abs(t_stat), len_df)*2
```

# item a: linear regression of `y` onto `x`:

```python
X = np.array(df_YontoX['X']).reshape(-1, 1)
Y = np.array(df_YontoX['Y'])

SLR_model_obj = linear_model.LinearRegression(fit_intercept=False)  # creating model object without Intercept
SLR_model_obj.fit(X,Y)

print( 'Intercept =', SLR_model_obj.intercept_ )
print( 'Beta in true f(x) = 2')
print( 'Beta from model =', round(SLR_model_obj.coef_[0],5) )
print( 'Beta from formula =', round(beta_approx(df_YontoX),5) )
print( 'STD beta from formula =', round(std_beta_approx(df_YontoX),5) )
print( 't-statistic from Y onto X formula =', round(t_stat_v1(df_YontoX),5) )
print( 't-statistic from symmetrical formula =', round(t_stat_v2(df_YontoX),5) )
print( 'p-value =', p_value(t_stat_v1(df_YontoX),len(df_YontoX)) )
print( 'R^2-statistics =', df_YontoX.corr().iloc[0,1]**2)
print( '_'*80,'\nDifference between sample Beta and real Beta makes',
       round( (beta_approx(df_YontoX) - 2) / std_beta_approx(df_YontoX), 2), 
       "beta's std (t-statistic for delta vs real beta)" )
```

## Conclusions:
1. The SLR model algorithm finishes Beta regression at exact minimum of RSS given by analytical formula.
2. The difference between Beta got from sample and real Beta ($\beta = 2$) **fits well** the sample beta std estimation.

## analytical explanation why slope for regression of `x` onto `y` is 0.4 instead of 0.5

The slope approximation from a sample is given by
$$ \mbox{slope} =\ \frac
{\sum_{i=1}^{n} x_i y_i}
{\sum_{i'=1}^{n} x_{i'}^2}
.$$

After exchanging X and Y in the initial $ y_i = \beta\ x_i + \epsilon_i $, we can use in the slope formula above $ y_i = \frac{x_i}{\beta} - \frac{\epsilon_i}{\beta} $:

$$ \mbox{slope} =\ \frac
{\sum_{i=1}^{n} x_i \Bigl(\frac{x_i}{\beta} - \frac{\epsilon_i}{\beta}\Bigr)}
{\sum_{i'=1}^{n} x_{i'}^2}
\ =\ \frac{1}{\beta} \Bigl( 1 - \frac{\sum_{i=1}^{n} x_{i} \epsilon_i}{\sum_{i'=1}^{n} x_{i'}^2} \Bigr).$$

If $\sum_{i=1}^{n} x_{i} \epsilon_i$ were zero, we would got intuitively expected $ \mbox{slope} = 1/\beta$, but as soon as we exchanges X and Y, now $\sum_{i=1}^{n} x_{i} \epsilon_i \ \ne\ 0$, they do correlate now.  To calculate it, let's again exchange X and Y in the notation (making now $\sum_{i=1}^{n} x_{i} \epsilon_i \ =\ 0$ !!):

$$
\mbox{slope} = \frac{1}{\beta} \Bigl( 1 - \frac{\sum_{i=1}^{n} (\beta\ x_{i} \epsilon_i + \epsilon_i^2)}{\sum_{i'=1}^{n} \beta^2 x_{i'}^2 + 2\beta x_{i'}\epsilon_{i'} + \epsilon_{i'}^2} \Bigr),\ \ \ \ \ \mbox{where} \sum_{i=1}^{n} x_{i} \epsilon_i \ =\ 0. 
$$

thus the relation between $\beta$ and the slope for reversed regression function **on the same sample**:

$$
\mbox{slope}_{rev} = \frac{1}{\beta} \Bigl( 1 - \frac{\sum_{i=1}^{n} \epsilon_i^2}{ \beta^2 \sum_{i'=1}^{n} x_{i'}^2  + \sum_{i'=1}^{n} \epsilon_{i'}^2} \Bigr)\ =\ 
\frac{1}{\beta} \Bigl( 1 - \frac{\mbox{var}(\epsilon)}{\beta^2\ \mbox{var}(x)  + \mbox{var}(\epsilon)} \Bigr).
$$

This is valid for any generated $y_i = \beta\ x_i + \epsilon_i$, when $x$ and $\epsilon$ are independent ($corr(x,\epsilon) \approx 0$).  In our particular case $X \sim N(0,1)$ and $\epsilon \sim N(0,1)$, therefore 
$$
\mbox{var}(x) = \mbox{var}(\epsilon) = 1, \ \ \ \ \ \ \ \mbox{slope}_{rev} = \Bigl(\beta + \frac{1}{\beta}\bigr)^{-1} ;
$$

$$
\mbox{if}\ \beta=2,\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \mbox{slope}_{rev} = 0.4\ , 
$$
that is exactly as we got it actually in our experiment !

## some considerations to calm down our intuition disorder about the results
And yet, we feel certain congnitive discomfort - even having mathematically explained the reason why reversed regression slope is not 1/2 but only 2/5 !  

The are two factors acting together to create the reason for this *mal-de-chien*:
1. We think that the line in the figure below is minimising the sum of squares of distance from each point of distribution to this line (that is orthogonal distance!), and if so, than the functional of the line position is completely symmetrical between X and Y, and the reversed slope should be exact $1/\beta$. While in fact optimisation is done for the sum of residual squares - that is of the distances measured parallel to predicted coords - in one case it is Y coord, in the reversed case it is X coord.
2. The way how we generated data means that "*statistically*" it always contains pairs of data points $(x,y)$ like { $(x,2x-\epsilon)$, $(x,2x+\epsilon)$  } - this pair has equal but opposit by sign residuals, but their leverage as compared to the pount $(0,0)$ is very different depending along which axis we take the residuals: for $(x,2x+\epsilon)$ the leverage for X onto Y regression is bigger than the one for $(x,2x-\epsilon)$, while if regression is done for Y onto X - both leverages are the same (see figure below left).

Therefore (1 and 2 being valid above) the optimal regression line for reversed regression (X onto Y) goes higher than regression line for (Y onto X)!

## Conclusions:
1. For each way of regression the stochastisity of regresor parameter makes slope less steep as compared to the slope of regular part of the dependence between parameters.
2. Actually if stochastisity of the parts is independent, we still are very close to the same formula for the $\mbox{slope}\ =\ f(\beta_{real})$.
3. We have inteentively selected sampling with stochastivity of each parameter equal to the variance of its opponent variable - to keep this formula simple (btw it is also symmetrical towards changing $\beta$ to $\beta^{-1}$):
$$ \mbox{slope}_{rev} = (\beta + \beta^{-1})^{-1} $$

```python
print('For regression of Y onto X: slope from sample = ', Beta_YontoX, 
      ', while from slope formula above =', 1/(0.5+2))

print('For regression of X onto Y: slope from sample = ', Beta_XontoY, 
      ', while from slope formula above =', 1/(2+0.5))
      
```

4. The results show, that even if synched part of X and Y has slope 2, by adding independent stochastik terms to both X and Y we can get same no-intercept-regression slope both for y(x) and x(y) by providing opposite balancing of stochastical terms - very discouraging result, considering the fact that actually X and Y always have some statictical errors in values.
5. Actually the formula in 6.1 can be written as follows:

$$
\mbox{For zero-intercept simple linear regression:}\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ 
\mbox{slope }_{YontoX}*\mbox{slope }_{XontoY}\ =\ {R^2} = corr(X,Y)^2
$$

```python
print('Beta_YontoX * Beta_XontoY =', round(Beta_YontoX*Beta_XontoY,5))
print('corr(X,Y)**2 =', round(df_XontoY.corr().iloc[0,1]**2,5))
```

6. BTW, for the initially generated data R^2 statistics was 0.8, while multiplication of two slopes for opposite regressions gave 2\*0.4=0.8 also!  Thus the form above looks like most universal and simple form of relations for two ways regressions: $$\mbox{slope }_{YontoX}*\mbox{slope }_{XontoY}\ =\ {R^2} = corr(X,Y)^2$$.
7. **The closer R^2 statistics between predictor and predicted perameters is to 1, the closer are the results to intuitive expectations to see $1/\beta$ in the reversed regression slope.**

# items d & e

Numerically we already confirmed that t-statistics (hence also p-values) are the same for both ways of regression - always!  Now let's derive it analytically:

$$ 
\frac{\hat{\beta}}{\mbox{SE}(\hat{\beta})}\ =\ 
\frac{ \sum_{i=1}^n x_i y_i } { \sum_{i=1}^n x_i^2 }\ \cdot\ \sqrt{
\frac{ (n-1)\ \sum_{i=1}^n x_i^2 } { \sum_{i=1}^n (y_i-x_i\hat{\beta})^2 }
}\ =\ 
\frac{\sqrt{n-1}\ \sum_{i=1}^{n} x_i y_i}
{
\sqrt{
\Bigl( \sum_{i=1}^n x_i^2 \Bigr)\cdot
\Bigl( \sum_{i=1}^n (y_i-x_i\hat{\beta})^2 \Bigr)
}
}
$$

and this should be equal to:

$$\frac
{\sqrt{n-1}\ \sum_{i=1}^{n} x_i y_i}
{\sqrt{
\Bigl(\sum_{i=1}^n x_{i}^2\Bigr)\ \Bigl(\sum_{i=1}^n y_i^2\Bigr)\ -\ \Bigl(\sum_{i=1}^{n} x_i y_i\Bigr)^2
}}.$$


Thus we need only to prove that

$$
\Bigl( \sum_{i=1}^n x_i^2 \Bigr)\cdot
\Bigl( \sum_{i=1}^n (y_i-x_i\hat{\beta})^2 \Bigr), \ \ \ \ \ \ \ \ \ \ \ \mbox{where}\ \  
\hat{\beta} = \frac{ \sum_{i=1}^n x_i y_i } { \sum_{i=1}^n x_i^2 }
$$

is equal to 

$$
\Bigl(\sum_{i=1}^n x_{i}^2\Bigr)\ \Bigl(\sum_{i=1}^n y_i^2\Bigr)\ -\ \Bigl(\sum_{i=1}^{n} x_i y_i\Bigr)^2
$$

Here we are:

$$
\Bigl( \sum_{i=1}^n x_i^2 \Bigr)\cdot
\Bigl( \sum_{i=1}^n (y_i^2-2y_i x_i\hat{\beta}+x_i^2\hat{\beta}^2) \Bigr)\ =\ 
\Bigl( \sum_{i=1}^n x_i^2 \Bigr)\cdot
\Bigl( \sum_{i=1}^n y_i^2 \Bigr)\ -\ 
\Bigl( \sum_{i=1}^n x_i^2 \Bigr)\cdot\hat{\beta}\cdot
\Bigl( \sum_{i=1}^n  x_i(2y_i-x_i\hat{\beta}) \Bigr)
$$

substituting one $\hat\beta$ with its expression above, and using definition of $\hat\beta$ as $y_i=\hat\beta\ x_i + \epsilon_i$, where $x_i$ and $\epsilon_i$ are statistically independent, we would have:

$$
\Bigl( \sum_{i=1}^n x_i^2 \Bigr)\cdot
\Bigl( \sum_{i=1}^n y_i^2 \Bigr)\ -\ 
\Bigl( \sum_{i=1}^n x_i y_i \Bigr)\cdot
\Bigl( \sum_{i=1}^n  x_i(y_i+\epsilon_i) \Bigr)\ =\ 
\Bigl( \sum_{i=1}^n x_i^2 \Bigr)\cdot
\Bigl( \sum_{i=1}^n y_i^2 \Bigr)\ -\ 
\Bigl( \sum_{i=1}^n x_i y_i \Bigr)^2\ -\ 
\Bigl( \sum_{i=1}^n x_i y_i \Bigr)\cdot
\Bigl( \sum_{i=1}^n  x_i \epsilon_i \Bigr)
,$$

where the last term is zero, because accoring to definition of $\hat\beta$ the $x_i$ and $\epsilon_i$ are independend and hence $\sum_{i=1}^n x_i \epsilon_i\ =\ 0$.  That's it !

## Conclusions:
1. Thus the formula for t-statistics is fully symmetrical for exchanging $x_i$ and $y_i$, hence for *any* sampling the t-statistics are the same in both ways of regressions (as well as R^2-statistics, being just a squared correlator between $x_i$ and $y_i$.) 

# item f: now the intercept is also defined by the sample
All is the same as in the first data sampling, but in the model the intercept is also found by optimization.  **NB!:** the analytical formulas of chapter 3 are no more valid - the parameters of f(x) are taken only from the fitted model.

```python
n_sampling = 100000
rng = np.random.default_rng()
X = rng.standard_normal(n_sampling)
Y = 2*X + rng.standard_normal(n_sampling)

df_YontoX = pd.DataFrame(
    [X,Y],
    index=['X','Y']
).T
df_XontoY = df_YontoX.rename(columns={'X':'Y','Y':'X'})
```

```python
X = np.array(df_YontoX['X']).reshape(-1, 1)
Y = np.array(df_YontoX['Y'])

SLR_model_obj = linear_model.LinearRegression()  # creating model object
SLR_model_obj.fit(X,Y)

print( 'Intercept =', SLR_model_obj.intercept_ )
print( 'Beta from model =', Beta_YontoX := round(SLR_model_obj.coef_[0],5) )
print( 'R^2-statistics =', df_YontoX.corr().iloc[0,1]**2)
```
